<h1>Normalized version of the data, plus adding three hidden layers, each of 10 nodes and ReLU activation function <h1>

```python
import pandas as pd
import numpy as np
```

Importing Concrete data for evaluation

```python
concrete_data = pd.read_csv('https://cocl.us/concrete_data')#importing the data using pandas
concrete_data.head() #Checking the first five items from the dataset
```

The data must be splited between the independent variable and dependent variable, that will be *(cement, blast furnance, fly slag,
water, superplasticizer, coarse aggregate, fine aggregate, and age)* and the dependent variable that will be *(Strenght)*

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

**Let's normalize the data of the predictors which is by subtracting the mean from the individual predictors and dividing by the standard deviation.**

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

We have to get the number of columns of the **predictors** variable that was **normalized** as the number of inputs to the deep learning model

```python
n_cols = predictors_norm.shape[1] # number of predictors
```

**Then we have to import keras library and as per video lesson, the model have two classes.**

```python
import keras

from keras.models import Sequential
from keras.layers import Dense
```

```python
# define classification model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))          
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

<h3> Now we have to randomly split a data into training data and 30% of the data randomly chosen as test data using 
    Scikit-learn.<h3> <h3> Therefore we have to import sklearn library to select train_test_split  <h3>
    <h2> Evaluating the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. Using the mean_squared_error function from Scikit-learn <h3>

**First let's import the mean_sqaure_error function from the scikit_learn library, Evaluating the model on the test data using "evaluate", and using 100 epochs**

```python
from sklearn.model_selection import train_test_split #importing the sklearn library for train_test_split
from sklearn.metrics import mean_squared_error#importing the mean_sqaure_error function from the sklearn library
```

```python
MSE=np.zeros(50)
for i in range(50):
    X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target, test_size=0.3, random_state=42) #test size is set to 30%
    model=regression_model() #Call the function to create a model
    model.fit(X_train, y_train, epochs= 50, verbose=2) #we then need to fit the x-train data along with y_train data and  fit it on a model
    print("Evaluate on test data")
    results=model.evaluate(X_test, y_test)
    print("test loss, test acc:", results) 
    target_predicted= model.predict(X_test)
    MSE[i]=mean_squared_error(y_test, target_predicted)
    print("Test set MSE for {} cycle:{}".format(i+1,MSE[i]))
        
```

**Let's get the mean and the standard deviation of the mean sqaured errors from previous cycle**

```python
mean_MSE = MSE[i].mean() #creating a variable reporting the mean value for the mean squared errors
std_MSE = MSE[i].std() #creating a variable reporting the standard value for the mean squared errors
```

```python
print("The mean of the mean squared errors is {}, and the standard deviation for the mean squared value is {}".format(mean_MSE, std_MSE))
```
